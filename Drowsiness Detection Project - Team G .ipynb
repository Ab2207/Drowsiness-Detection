# Extracting frames from videos and saving it to "drowiness_data" file 

```python
# Importing all necessary libraries 
import cv2 
import os 
  
# Read the video from specified path 
cam = cv2.VideoCapture("/Users/AB/Desktop/0.mp4") 
  
try: 
      
    # creating a folder named data 
    if not os.path.exists('drowsiness_data'): 
        os.makedirs('drowsiness_data') 
  
# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of data') 
  
# frame 
currentframe = 0
  
while(True): 
      
    # reading from frame 
    ret,frame = cam.read() 
  
    if ret: 
        # if video is still left continue creating images 
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name) 
  
        # writing the extracted images 
        cv2.imwrite(name, frame) 
  
        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break
  
# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```

# Data cleaning 
# 1. Color to Grayscale conversion
# 2. Standard size conversion 
# 3. Cropping image to focus on face area


```python
import cv2
import os
```

1. Color to Grayscale conversion

```python
def loadImages(path):
   return [os.path.join(path, f) for f in os.listdir(path) if f.endswith(".jpg")]

filenames = loadImages("/Users/AB/Desktop/Sample_data")
images = []
for file in filenames: 
    f = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    images.append(f)
```

2. Resize Images 

```python
import numpy as np 
images = np.array(images)
```

```python
w, h = 1000, 800
resized_images = [cv2.resize(img, (w,h), interpolation=cv2.INTER_LINEAR) for img in images] 
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(resized_images[30], cmap = 'gray')
plt.grid(False)
plt.show()
```

3. Cropped Images to focus on the face

```python
cropped_images = []

for i in resized_images: 
    imgCropped = i[150:600, 500:1000]
    cropped_images.append(imgCropped)
cropped_images = np.array(cropped_images)
cropped_images.shape
```

```python
plt.figure()
plt.imshow(cropped_images[30], cmap = 'gray')
plt.grid(False)
plt.show()
```
